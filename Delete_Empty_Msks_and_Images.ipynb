```python
# This code deletes (removes) all masks and the corresponding image chips that are only black (only one color)

from PIL import Image
import os
path = "path to masks/"
path2 = "path to images/"

count = 0

for filename in os.listdir(path):
    img = Image.open(path + '\\' + filename)
    clrs = img.getcolors()
    if len(clrs) == 1:
        count = count + 1
        print(filename, len(clrs), count)
        os.remove(path + '\\' + filename)
        os.remove(path2 + '\\' + filename)
print("number of deleted images:", str(count))
print("number of deleted masks:", str(count))

```
